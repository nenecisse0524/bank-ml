# 4-a: Let's concat first and last name.




```python
def concat_names(row):
    first_name = row['nameFirst']
    last_name = row['nameLast']
    return f"{first_name} {last_name}"
```

```python
import pandas as pd

data = '/content/drive/MyDrive/Baseball dataset/Master.csv'
# Try reading the file with a different encoding, like Latin-1
df = pd.read_csv(data, encoding='latin-1')  # Or try 'iso-8859-1'

df['full_name'] = df.apply(concat_names, axis=1)
print(df)

# Use the existing column names 'nameFirst' and 'nameLast'
df['full_name'] = df['nameFirst'] + ' ' + df['nameLast']
print(df)
```

# 6-c : now we are going to make a histogram of triples (3B) per year.

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Baseball dataset/Batting.csv')
df.head(10)
triples_by_year = df.groupby('yearID')['3B'].sum()
```

```python
plt.hist(triples_by_year, bins=20)
plt.xlabel('Year')
plt.ylabel('Number of 3B')
plt.title('Number of 3B per Year')
plt.show()
```
